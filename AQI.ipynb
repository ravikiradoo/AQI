```python
from bs4 import BeautifulSoup
import pandas as pd
Data=[]
for year in range(2013,2019):
    for month in range(1,13):
        data=open("./DATA/HTML/{}/{}.html".format(year,month),"rb")
        soup = BeautifulSoup(data,'html.parser')
        results = soup.find('table',attrs={'class':'medias mensuales numspan'}).find_all('tr')
        for row in results:
            ROW=[col.text for col in  row.find_all('td')]
            if ROW and ROW[0].isnumeric():
                ROW[0] =  ROW[0] + '/' + str(month) + '/' +str(year)
                Data.append(ROW)
```

```python
header = []
data=open("./DATA/HTML/header.html","rb")
soup = BeautifulSoup(data,'html.parser')
header = [ h.text for h in soup.find('tr').find_all('th')]
```

```python
DF = pd.DataFrame(Data,columns=header)
```
